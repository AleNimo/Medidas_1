```python
import numpy as np
```

# Estudio Tipo A

```python
V_mean = 100.145
V_ui = 1.489 #desvio muestral
N = 20 #Muestras

print("Incertidumbre tipo A de %0.4f" %V_ui)
```

# Estudio Tipo B
## Datos del multímetro:
### 3 y 1/2 dígitos

### Error máximo = +/-(0.5% lectura + 3 dígitos)

### Entonces la medicion hecha con el multímetro en el rango de 200V hace que, si el último digito puede variar +/- 3, haya un error de +/-0.3V además del porcentual de 0.5%.

```python
error_porcentual = 0.5
error_absoluto_digito = 0.3

error_absoluto = error_porcentual/100 * V_mean + error_absoluto_digito

#Supongo distribución uniforme del error_absoluto
V_uj = error_absoluto / np.sqrt(3)

print("Error absoluto = %0.4f" %error_absoluto)
print("Incertidumbre tipo B = %0.4f" %V_uj)
```

# Cálculo de Incertidumbre Combinada

```python
#Al ser una medición directa, la derivada parcial es 1, y no hay análisis de correlación.
V_uc = np.sqrt(V_ui**2 + V_uj**2)

#Analizo distribución de la incertidumbre combinada
Veff = V_uc**4 / (V_ui**4 / (N-1))

print("Incertidumbre combinada = %0.4f" %V_uc)
print("Grados de libertad efectivos = %0.4f" %Veff)
```

## Incertidumbre expandida
### Como la incertidumbre tipo A es mayor a la tipo B, y eso concuerda con que los grados efectivos no son muy altos (casi 22 parecido a los 19 grados de libertad del tipo A) concluimos que la distribución de la incertidumbre combinada es t-student con 22 grados de libertad.

### La tabla que indica la constante de expansión para asegurar un porcentaje de confianza determinado, indica valores para 20 grados efectivos y luego 25.
### Uso la fila de 20 para que la incertidumbre se expanda de más y no de menos.

```python
V_uc_expandida = V_uc * 2.09 #(para Veff = 20 y 95% de confianza)

print("La medición es de %0.3fV" %V_mean,"con una incertidumbre de +/-%0.4fV" %V_uc_expandida, "y con una confianza del 95%")
```
